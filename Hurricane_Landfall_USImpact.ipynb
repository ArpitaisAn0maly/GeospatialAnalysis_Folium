```python
#adding year column



df1['date'] = pd.to_datetime(df1['date'].astype(str), format = '%Y %m %d')

df1['year'] = df1['date'].map(lambda x: x.year)

df1
```

```python
booleans =[]

for date in df1['year']:
    if date > 1999:
        booleans.append(True)
    else: 
        booleans.append(False)
date_range = pd.Series(booleans)
new_df1 = df1[date_range]

new_df1.shape
new_df1 #filtered by date 2000-2020
```

```python
dff1= new_df1[(new_df1['record_id']=='L')]
```

```python
#drop duplicates


filtered_df1 = dff1.sort_values(by='time', ascending=False)
filtered_df1 =filtered_df1.drop_duplicates(subset='name', keep="first")
filtered_df1 = filtered_df1.sort_values(by ='year', ascending = True)
```

```python
mergeddf = pd.merge(dff,filtered_df1,how='left', on=["storm_id"])
```

```python
mergeddf1=mergeddf[['storm_id', 'name_x', 'records_x', 'landfall', 'hurricane',
       'hurr_landfall', 'lifetime_max_wind', 'lifetime_min_pres',
       'init_latitude', 'init_longitude', 'init_date', 'final_latitude',
       'final_longitude', 'final_date', 'flf_latitude', 'flf_longitude',
       'flf_date', 'status']].copy()
```

```python
mergeddf1['status'] = mergeddf1['status'].fillna('HU')
```

```python
mergeddf1["flf_date"]  = pd.to_datetime(mergeddf1["flf_date"] .astype(str), format = '%Y %m %d')
```

```python
# landfall_map = folium.Map(zoom_start=5,location=[37.0902, -95.7129])

def color_producer(lifetime_max_wind):
    if lifetime_max_wind < 111.00:
        return 'green'
    elif 111 <= lifetime_max_wind < 129:
        return 'orange'
    elif 130 <= lifetime_max_wind < 156:
        return 'red' 
    else:
        return 'darkpurple'
      
def cat_producer(lifetime_max_wind):
    if lifetime_max_wind < 111.00:
        return 'Cat 2'
    elif 111 <= lifetime_max_wind < 129:
        return 'Cat 3'
    elif 130 <= lifetime_max_wind < 156:
        return 'Cat 4' 
    else:
        return 'Cat 5'
      
for i, r in mergeddf1.iterrows():
        
    #setting for the popup
    popup=folium.Popup('Name : '+str(r['name_x'])+'\n'+'Status : '+str(r['status'])+ ':-' +str(cat_producer(r['lifetime_max_wind']))+'\n'+'Date of Landfall : '+str(r['flf_date']),max_width=800)
       
    #Plotting the Marker for each landallト
    folium.map.Marker(
        location=[r['flf_latitude'], r['flf_longitude']], 
        popup=popup ,
        icon=folium.Icon(color=color_producer(r['lifetime_max_wind']),icon="spinner", prefix='fa')
    ).add_to(Landfall)
    
Landfall
```
